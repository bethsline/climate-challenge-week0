```python
# Cell 1: Load and initial inspection
df = pd.read_csv(r"C:\Users\usre\Desktop\10 Academy\data\ethiopia.csv")

# Add country column
df['Country'] = 'Ethiopia'

print(f"Shape: {df.shape}")
print(f"\nMissing values (before cleaning):")
print(df.isna().sum())
```

```python
# Cell 2: Replace -999 with NaN
df = df.replace(-999, np.nan)

print("After replacing -999:")
print(df.isna().sum())
print(f"\nPercentage missing:")
print((df.isna().sum() / len(df) * 100).round(2))
```

```python

df['Date'] = pd.to_datetime(df['YEAR'].astype(str) + 
                            df['DOY'].astype(str).str.zfill(3), 
                            format='%Y%j')
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

print(f"Date range: {df['Date'].min()} to {df['Date'].max()}")
print(df[['YEAR', 'DOY', 'Date', 'Month', 'Year']].head())
```

```python

duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

if duplicates > 0:
    df = df.drop_duplicates()
    print(f"Removed {duplicates} duplicates")
```
